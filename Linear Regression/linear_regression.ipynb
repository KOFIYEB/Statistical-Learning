```r
#loading data
data <- read.table("USA_Housing.csv", sep = ",", header = TRUE)
```

```r
#attributing the variables
Price <- data$Price    # y variable
Avg..Area.Income <- data$Avg..Area.Income # x1
Avg..Area.House <- data$Avg..Area.House  # x2
Avg..Area.Number.of.Rooms<- data$Avg..Area.Number.of.Rooms   # x3
Avg..Area.Number.of.Bedrooms <- data$Avg..Area.Number.of.Bedrooms # x4
Area.Population <- data$Area.Population # x5
```

```r
#PLOTTING THE LINEAR REGRESSION
M1 = lm(Price ~ Avg..Area.Income); M2 = lm(Price ~ Avg..Area.House );
M3 = lm(Price~ Avg..Area.Number.of.Rooms) ; 
M4 = lm(Price~ Avg..Area.Number.of.Bedrooms );M5 = lm(Price~ Area.Population)
#par(mfrow = c(1, 5))
plot(Avg..Area.Income, Price, pch = 1)
abline(M1, col = "red")
title("PRICE REGRESSION TO AVG. AREA INCOME")
```

```r
plot(Avg..Area.House ,Price, pch = 1)
abline(M2, col = "red")
title("PRICE REGRESSION TO AVG. AREA HOUSE")
```

```r
plot(Avg..Area.Number.of.Rooms, Price, pch = 1)
abline(M3, col = "red")
title("PRICE REGRESSION TO AVG. AREA NUM OF ROOMS")
```

```r
plot(Avg..Area.Number.of.Bedrooms, Price, pch = 1)
abline(M4, col = "red")
title("PRICE REGRESSION TO AVG. AREA NUM OF BEDROOMS")
```

```r
plot(Area.Population, Price, pch = 1)
abline(M5, col = "red")
title("PRICE REGRESSION TO AREA POPULATION")
```

```r
# SIMPLE LINEAR REGRESSION
#All models are statistically significant (p value is smaller than 0.05)
model1 <- lm(Price ~ Avg..Area.Income, data = data)
summary(model1)
```

```r
model2 <- lm(Price ~ Avg..Area.House, data = data)
summary(model2)
```

```r
model3 <- lm(Price ~ Avg..Area.Number.of.Rooms, data = data)
summary(model3)
```

```r
model4 <- lm(Price ~ Avg..Area.Number.of.Bedrooms, data = data)
summary(model4)
```

```r
model5 <- lm(Price ~ Area.Population, data = data)
summary(model5)
```

```r
#MULTIPLE LINEAR REGRESSION
model6 <- lm(Price ~ Avg..Area.Income + Avg..Area.House+ Avg..Area.Number.of.Rooms
+ Avg..Area.Number.of.Bedrooms+ Area.Population , data = data )
summary(model6)
```

```r
#relationship between the response and predictors USING F STATISTICS
y_mean = mean(Price)
tss = sum((Price - y_mean)^2)
rss = sum((Price - model6$fitted.values)^2)
f = ((tss - rss) / 5)/(rss/(5000-5-1))
f
```

```r
#The large F-statistic suggests that at lease one of the advertising media must be related to Profits.
#Therefore, we can reject H0
# access f test score from summary()
summary(model6)$fstatistic[1]
```

```r
# Split the data into training and testing sets
set.seed(123)
data_without_last_col <- data[, -ncol(data)]
splitIndex <- createDataPartition(data_without_last_col$Price, p = 0.8, list = FALSE)

trainData <- data_without_last_col[splitIndex,]
testData <- data_without_last_col[-splitIndex,]
```

```r
# Fit the multi linear regression model on the training data
model8 <- lm(Price ~ ., data = trainData)
```

```r
# Predict the values on the test data
predictions <- predict(model8, testData)
```

```r
# Calculate the mean error CV
meanErrorCV <- mean(abs(testData$Price - predictions))
print(meanErrorCV)
```
